```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

history_path = r".\history.csv"

df = pd.read_csv(history_path)

print(df.head())
print(df.columns)
```

```python
plt.figure(figsize=(6, 4))

plt.semilogy(df["epoch"], df["train_mse"], label="Train MSE")

# val_mse 有些 epoch 可能是 NaN，只画有效值
valid = df["val_mse"].notna()
plt.semilogy(df.loc[valid, "epoch"], df.loc[valid, "val_mse"], label="Validation MSE")

plt.xlabel("Epoch")
plt.ylabel("MSE loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))

plt.semilogy(df["epoch"], df["train_rel_l2"], label="Train relative L2")

valid = df["val_rel_l2"].notna()
plt.semilogy(df.loc[valid, "epoch"], df.loc[valid, "val_rel_l2"], label="Validation relative L2")

plt.xlabel("Epoch")
plt.ylabel("Relative L2 error")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
